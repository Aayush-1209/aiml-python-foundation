```python
import matplotlib.pyplot as plt
```

```python
# Task
days = ["Mon", "Tue", "Wed", "Thu", "Fri"]

cities = ["New York", "London", "Delhi", "Tokyo"]

temperatures = [
[22, 23, 21, 24, 25], #New york
[18, 19, 17, 20, 21], # London
[30, 32, 31, 33, 34], # Delhi
[25, 26, 24, 27, 28]  #Tokyo
]
fig , axes = plt.subplots(2,2)
city_no = 0
for i in range(2): #rows (i) - 0,1
    for j in range(2):#rows (j) - 0,1
        axes[i][j].plot(days,temperatures[city_no],marker='o')
        axes[i][j].set_title(cities[city_no])
        city_no+=1
fig.suptitle("Temperature in cities over week")
fig.supylabel("Temp")
fig.supxlabel("Days")
fig.tight_layout()
```
