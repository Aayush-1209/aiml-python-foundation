```python
#Seaborn - Datavisualisation lib based on matplotlib

# In built data sets ,  high level features
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# sns.get_dataset_names()

tips = sns.load_dataset("tips")
print(type(tips))
print(tips.head())
```

```python
sns.set_theme(style='darkgrid')
sns.relplot(
    data = tips,
    x = "total_bill",
    y="tip",hue='smoker',size='size',style='sex'
)
```

```python
sns.relplot(
    data = tips,
    x = "total_bill",
    y="tip",kind='line'
)
```

```python
# Heatmap
flights = sns.load_dataset("flights")
flights_pivot = flights.pivot(index="month", columns="year", values="passengers")
print(flights_pivot)

sns.heatmap(
    flights_pivot,
    cmap="coolwarm", annot=True, fmt="d"
)
plt.title("Heatmap for Passengers")
plt.tight_layout()
plt.show()
```

```python
# Relational plots
# scatter plot
sns.get_dataset_names()
tips = sns.load_dataset("tips")
print(tips.head())

sns.scatterplot(
    data=tips,
    x="total_bill",
    y="tip",
    hue="time"
)
plt.show()

# line plot
flights = sns.load_dataset("flights")

sns.lineplot(
    data=flights,
    x="year",
    y="passengers"
)
plt.show()
```

```python
# Bar plot
sns.barplot(
    data=tips,
    x="day",
    y="total_bill",
    hue="sex"
)
plt.show()

# Box plot
sns.boxplot(
    data=tips,
    x="day",
    y="total_bill"
)
plt.show()
```

```python
# histogram
penguins = sns.load_dataset("penguins")

sns.histplot(
    data=penguins,
    x="body_mass_g",
    bins=30
)

print(penguins.head())
plt.show()
```

```python
# Recommended approach
fig, ax = plt.subplots()

sns.lineplot(
    data=tips,
    x="day",
    y="total_bill",
    hue="sex",
    marker="o",
    errorbar=None,
    ax=ax
)

ax.set_title("Average Total Bill by Day and Sex")
ax.set_xlabel("Day of Week")
ax.set_ylabel("Total Bill ($)")

ax.legend(title="Sex")
plt.tight_layout()

plt.show()
```
