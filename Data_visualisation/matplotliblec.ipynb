```python
#Oscar winning movies
oscar_movies = [
"The Dark Knight",
"The Hurt Locker",
"The King's Speech",
"The Artist",
"Argo"]

years = [2008, 2009, 2010, 2011, 2012]
oscar_revenue = [1005, 170, 427, 133, 232] # in $M
#plot
plt.plot(years,oscar_revenue)
plt.title("Oscar movies rev. in each year")
plt.xlabel("years")
plt.ylabel("Revenue in ($M)")
```

```python
# Non-Oscar Winner Movies
non_oscar_movies = ["Slumdog Millionaire","Avatar","Inception","Hugo","Lincoln"]
non_oscar_revenue = [378, 2788, 829, 185, 275] # in $M

plt.plot(years,oscar_revenue,"o--y",label="Oscar movies")
plt.plot(years,non_oscar_revenue,color="red",marker="o",linestyle="-.",linewidth=4,label="Non-Oscar movies")

plt.title("Oscar vs Non Oscar movies revenues")
plt.xlabel("Years")
plt.ylabel("Revenue in $M")

plt.legend(loc="upper right")   # default = "upper right" 

#Format strings = consists of a part for color , marker and line : fmt = '[marker][line][color]'

```

```python
#Vertical Bar chart
movies = [
"The Dark Knight",
"The Hurt Locker",
"The King's Speech",
"The Artist",
"Argo"]

years = [2008, 2009, 2010, 2011, 2012]
revenue = [1005, 170, 427, 133, 232] # in $M

plt.bar(years,revenue,color="green")
plt.title("Rev in each year bar")
plt.xlabel("Year")
plt.ylabel("Rev")
#To plot exact value in the bar , or in cordinates use plt.text(x-axis,y-axis,text,allignment)
for i in range(len(years)):
    plt.text(years[i],revenue[i]+10,str(revenue[i]),ha="center")
    
plt.ylim(0,max(revenue)+100)#to extend the graph in y axis
plt.xlim(2007,max(years)+1) #to extend the graph in x axis 
```

```python
import numpy as np

#BarChart compare 2 sets
oscar_movies = [
"The Dark Knight",
"The Hurt Locker",
"The King's Speech",
"The Artist",
"Argo"]

years = [2008, 2009, 2010, 2011, 2012]
oscar_revenue = [1005, 170, 427, 133, 232] # in $M

non_oscar_movies = ["Slumdog Millionaire","Avatar","Inception","Hugo","Lincoln"]
non_oscar_revenue = [378, 2788, 829, 185, 275] # in $M

# plt.bar(years,oscar_revenue)
# plt.bar(years,non_oscar_revenue)
#if we plot like above it'll overlap to get it proper we have to set the width
# 1-Bar = width is 0.8(Default) value and range in x-axis is 0-1(Default) 

x = np.arange(len(years))
width =0.4
plt.bar(x-width/2,oscar_revenue,width,label="Oscar Movies")
plt.bar(x+width/2,non_oscar_revenue,width,label="Non-oscar Movies")

plt.title("Oscar vs non-oscar movies revenue")
plt.xlabel("Years")
plt.ylabel("Revenue in $M")

plt.legend()

plt.xticks(x,years)
```

```python
#Scatter plot 

people = ["Person A", "Person B", "Person C", "Person D", "Person E",
"Person F", "Person G", "Person H", "Person I", "Person J"]
age = [22, 25, 30, 35, 40, 45, 50, 55, 60, 65]
bp = [110, 115, 120, 122, 125, 130, 135, 123, 145, 150]

colors = ["green" if x<135 else "red" for x in bp]
# sizes = s , alpha for transparency 0(invisible) to 1 (visible)
#Color map =  cmap (color range )  ,  c desides which color point = which marker 
# plt.scatter(age,bp,color=colors,s=bp,alpha=0.6)
plt.scatter(age,bp,s=bp,cmap="OrRd",c=bp) #For color map
plt.title("Age vs bp")
plt.xlabel("Age")
plt.ylabel("BP")

plt.grid()
plt.colorbar(label="BP")

#Annotation which value is for which person
for i in range(len(people)):
    plt.annotate(people[i],xy=(age[i],bp[i]),xytext=(age[i]+1,bp[i]-1)) #xytext is for position

plt.xlim(min(age),max(age)+10)
plt.ylim(min(bp),max(bp)+10)
```

```python
#Pie charts

expenses = ["salaries","rent","marketing","r&d","miscellaneous"]

amount = [500,150,120,100,50]
explode = [0,0,0,0.2,0]
plt.pie(amount,labels=expenses,autopct="%1.1f%%",
        wedgeprops={
            "edgecolor":"black",
            "linewidth":1,
            "linestyle":"--",
            "alpha":0.7
        },explode=explode,startangle=90,shadow=True
        
       
       )
# plt.style.available  - all the available styles but use before plot

```

```python
# PIE chart Customizations

#  autopct="%1.1f%%" - shows percentage of each slice

#  colors - add custom colours

#  Wedgeprops - define (edgecolor, linewidth ,linestyle, alpha, fill ... )

#  explode - highlights a slice by pulling it out

#  startangle - rotate the chart

# . shadow - adds 3D effect or makes the chart circular
```

```python
# multiple datasets
legit_transactions = [
2.99, 5.49, 8.99, 12.50, 14.99, 19.99, 23.45, 29.99, 34.99, 39.50,
45.00, 49.99, 55.25, 60.00, 75.99, 89.99, 120.50, 150.00, 199.99,
249.99, 300.75, 450.00, 600.00, 850.00, 1200.00]

fraud_transactions = [
50, 100, 150, 200, 300, 500, 500, 750, 1000, 1000,
1200, 1500, 1500, 2000, 2500, 3000, 3000]

# Transaction Amount Distrbution - Legit vs Fraud
plt.hist(legit_transactions,color="green",bins=10,label="legit",edgecolor="black",alpha=0.5)
plt.hist(fraud_transactions,color="red",bins=10,label="fraud",edgecolor="black",alpha=0.5)

plt.title("Trans. legit vs fraud")
plt.xlabel("Amount in rs")
plt.ylabel("Frequency")
plt.legend()

# axvlines
```

```python
#Box plot
# A box plot is a statistical visualization that summarizes the distribution of a
# dataset using five key numbers:
#  Minimum (lowest non-outlier)
#  First Quartile (Q1) - 25th percentile
#  Median (Q2) - 50th percentile
#  Third Quartile (Q3) - 75th percentile
#  Maximum (highest non-outlier)
data = [3,4,5,6,7,8,9,13,15,20]
plt.boxplot(data)
plt.grid()
```

```python
#Stack plot 
# Stack Plot
days = ["Mon", "Tue", "Wed", "Thu", "Fri", "Sat", "Sun"]
direct = [50, 60, 70, 80, 90, 100, 110]
organic = [30, 40, 50, 55, 60, 70, 80]
social = [20, 25, 30, 35, 40, 50, 60]

plt.stackplot(days,direct,organic,social,labels=["Direct", "Organic" ,"Social"])
plt.title("Marketing dat aof the week")
plt.xlabel("Days")
plt.ylabel("# of customers")
plt.legend()
```
