```python
import requests
from bs4 import BeautifulSoup
```

```python
# URL = 'https://quotes.toscrape.com/page/1/'
page_count = 1

while True:
    URL = f"https://quotes.toscrape.com/page/{page_count}/"
    res = requests.get(URL)

    soup = BeautifulSoup(res.text , "lxml")
    quotes = soup.select("div.quote")
    # print(quotes)
    if not quotes:
        print("No valid pages")
        break
    with open(f"scraped_data/quotes{page_count}.html" , "wb") as f:
        f.write(res.content)
        print(f"Successfully downloaded the page {page_count}")

    page_count = page_count + 1
        
    
```

```python
all_life_quotes = []
file_count = 1

while True:
    try:
        with open(f"scraped_data/quotes{file_count}.html", "rb") as f:
            html_content = f.read()

        soup = BeautifulSoup(html_content, "lxml")
        all_quotes = soup.select("div.quote")

        for q in all_quotes:
            all_tags = [tag.get_text() for tag in q.select(".tags .tag")]

            if "life" in all_tags:
                text = q.select_one("span.text").get_text(strip=True)
                author = q.select_one("small.author").get_text(strip=True)
                all_life_quotes.append([text, author])

        file_count += 1

    except FileNotFoundError:
        print("No more files found. Stopping.")
        break
```
