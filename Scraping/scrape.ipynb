```python
import requests
```

```python
url = 'https://www.scrapethissite.com/pages/simple/'

res = requests.get(url)
```

```python
with open("scraped_data/data1.html","wb") as f:
    f.write(res.content)
```

```python
# Beautifulsoup
from bs4 import BeautifulSoup

with open("scraped_data/data1.html","r") as f:
    html_content = f.read()
soup = BeautifulSoup(html_content, "lxml")
    
```

```python
# soup.find(),find_all(),find_parent(),find_next(),find_previous()
```

```python
soup.find('h1')  #first occurence
soup.h1    #first occurence
```

```python
all_h3 = soup.find_all('h3')
## css selectors select() , select_one()

all_countries = []
for h3 in all_h3:
    name = h3.get_text(strip=True)
    # print(h3.find_parent("div")['class']) #find parent of particular tag and ['class'] this is to print all class names
    # print(h3.find_next("div").select("span.country-population")[0].get_text(strip=True)) # print the next sibling
    # # and selecting the tag with some class select returns the list 
    population = h3.find_next("div").select_one("span.country-population").get_text(strip=True)
    # by attributes 
    # print(h3.attrs)
    all_countries.append([name,population])
```

```python
# all_countries

import pandas as pd 

df = pd.DataFrame(all_countries,columns=["Name" , "Population"])
```

```python
df.to_csv("cleaned_data/data.csv",index=False)
```
